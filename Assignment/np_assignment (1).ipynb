```python
# array of num
import numpy as np
arr=np.arange(1, 11)
print(arr)
```

```python
# 3x3 array of 0
import numpy as np
arr=np.zeros((3, 3))
print(arr)
```

```python
# 4x4 array
import numpy as np
arr=np.full((4,4),7)
print(arr)
```

```python
# array of even numbers
import numpy as np
arr=np.arange(10, 31)
even=arr[arr % 2==0]
print(even)
```

```python
# identity matrix
import numpy as np
arr=np.eye(5,5)
print(arr)
```

```python
# shape, size, datatype
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])

print("shape is: ", arr.shape)
print("size is", arr.size)
print("data type is", arr.dtype)
```

```python
# reshape array
arr=np.arange(1,13)
new=arr.reshape(3,4)
print(new)
```

```python
# greater than 50
arr = np.array([10, 60, 45, 80, 30, 90])
new=arr[arr>50]
print(new)
```

```python
# mean, median, std
arr = np.array([5, 10, 15, 20, 25])
print("mean is: ", np.mean(arr))
print("median is", np.median(arr))
print("std is: ", np.std(arr))
```

```python
# adding two arrays
a = np.array([1,2,3])
b = np.array([4,5,6])
print("sum is", a+b)
print("sum is", np.add(a,b))
```

```python
# max, min in random arr
arr=np.random.rand(3,3)
print(arr)
print("max value is: ", np.max(arr))
print("min value is", np.min(arr))
```

```python
# sorting array
arr = np.array([12, 4, 7, 1, 9])
print("sorted array is:", np.sort(arr))
```

```python
# replacing all -ve num
arr = np.array([3, -1, 5, -7, 8])
arr[arr < 0]=0
print("replaced: ", arr)
```
