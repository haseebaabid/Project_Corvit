```python
from google.colab import drive
drive.mount("/content/drive")

import pandas as pd

file_path="/content/drive/MyDrive/StudentDataset.csv"

df=pd.read_csv(file_path)

print(df.head(5))

df.info()

```

```python
values=df.isnull().sum()
print(values)

df["null_count"]=df.isnull().sum()

print(df.head())
```

```python
dropped_colmn=df.dropna(how="all", axis=1)
print(dropped_colmn)
```

```python
print(df.ffill())
```

```python
print(df.fillna(2))
```

```python
print("marks greater than 85")
marks=df[df["Marks"]>85]
print(marks)

print("students from Lahore")
students=df[(df["City"]=="Lahore") & (df["Attendance"]>85)]
print(students)

print("third condition")
stu=df[(df["ProjectScore"]>85) & (df["Marks"]>80)]
print(stu)
```

```python
df["Marks"]+=5
print(df)
```

```python
df.loc[df["Marks"]<80, "Marks"]+=5
print(df)
```

```python
df["total_score"]=df["Marks"]+ df["ProjectScore"]
print(df)
```

```python
df["Performance_index"]=df["Marks"] * df["Attendance"]
print(df)
```

```python
marks=df.groupby("Department")["Marks"].mean()
print(marks)
```

```python
city=df.groupby("City")["Name"].count()
print(city)
```

```python
marks=df.groupby("Department")["Marks"].agg(["mean", "max", "min"])
print(marks)
```

```python
marks=df.groupby("City").agg(
    {
        "Attendance":"mean",
        "ProjectScore":"max"
    }
)
print(marks)
```
