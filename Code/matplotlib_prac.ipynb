```python
import pandas as pd
import matplotlib.pyplot as plt

info={
    "name":["sara", "zara", "ali", "hania", "sana"],
    "age":[21,31,12,34,19],
    "city":["isb", "lhr", "isb", "kchi", "bwp"]
}

df=pd.DataFrame(info)

plt.bar(df["name"], df["age"])
plt.xlabel("name of people")
plt.ylabel("age of people")
plt.title("age of people")
plt.legend(["Legend"])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

info={
    "name":["sara", "zara", "ali", "hania", "sana"],
    "age":[21,31,12,34,19],
    "city":["isb", "lhr", "isb", "kchi", "bwp"]
}

df=pd.DataFrame(info)

plt.bar(df["city"], df["age"])
plt.title("bar graph")
plt.xlabel("names")
plt.ylabel("age")
plt.legend(["legend"])
```

```python
import matplotlib.pyplot as plt

x=[12,18,30,20]
y=[31,12,32,1]

plt.plot(x, y, color="maroon", marker="o")
plt.title("basic graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(["legend"])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: months and sales
months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun"])
sales = np.array([1700, 2200, 2500, 2700, 3000, 3200])

plt.figure(figsize=(6,6))
plt.plot(months, sales, color="pink", marker="o", linestyle="-", linewidth=2, markersize=12)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("my plot")
plt.grid(True)
plt.show()
```
