```python
new=np.array([11,22,33,44])

reshaped=(new.reshape(2,2))
print("reshaped array is: ", reshaped)

transposed=reshaped.transpose()
print(transposed)

new.sort()
print(new)

max_index=reshaped.argmax()
print(max_index)

min_index=reshaped.argmin()
print(min_index)

```

```python
first=np.array([1,2,3,4,5])
second=np.array([1,4,3,8,5])

print(first + second)
print(first - second)
print(first * second)
print(first // second)
print(first % second)


print(first == second)
print(first != second)
print(first > second)
print(first <= second)
```

```python
first=np.array([1,2,3,4,5])
second=np.array([1,4,3,8,5])

#searching
new=np.where(first==2)
print("where result gives index", new)

# concatenate
next=np.concatenate((first, second))
print(next)

# filter
n=first[first > 1]
print(n)

ne=np.exp2(first)
print(ne)
```
