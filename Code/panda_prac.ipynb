```python
# Use of apply() to apply custom function
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank','Ali','Zia'],
    'City': ['NY', 'LA', 'NY', 'LA', 'NY', 'LA','BWP','BWP'],
    'Age': [24, 30, 22, 35, 28, 40, 12,15]
}
df = pd.DataFrame(data)


def custom_agg(group):
  return {"min_age": df["Age"].min(), "max_age":group["Age"].max(), "mean_age":group["Age"].mean()}

grouped=df.groupby("City").apply(custom_agg)
print(grouped)
```
