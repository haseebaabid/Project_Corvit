```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn. preprocessing import LabelEncoder

from google.colab import drive
drive.mount("/content/drive")
file_path="/content/drive/MyDrive/credit.csv"

df=pd.read_csv(file_path)

# print(df.head())
# print(df.value_counts())

print(df.isnull().sum())

df=df.drop(columns=["transaction_id", "transaction_hour", "cardholder_age"])

col_to_convert=df.select_dtypes(include=["bool", "object", "category"]).columns

if "is_fraud" in col_to_convert:
  col_to_convert=col_to_convert.drop("is_fraud")

df=pd.get_dummies(df, columns=col_to_convert, drop_first=True)

x=df.drop("is_fraud", axis=1)
y=df["is_fraud"]

x,y=x.align(y, join="inner", axis=0)

x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3)

print(x_train.shape)
print(x_test.shape)

# print(df.nunique())

model=RandomForestClassifier(
    n_estimators=5,
    min_samples_leaf=1,
    min_samples_split=5,
    max_depth=10,
    criterion="gini",
    random_state=43
)
model.fit(x_train, y_train)
features=x.columns

y_pred=model.predict(x_test)

accuracy=accuracy_score(y_test, y_pred)
print(f"accuracy is {accuracy}")

classification=classification_report(y_test, y_pred)
print(f"classification report: {classification}")

confusion=confusion_matrix(y_test, y_pred)
print(f"confusion matrix is: {confusion}")

imp=model.feature_importances_

for feat, importance in zip(features, imp):
  print(f"{feat}: {importance}:.2f")
```
