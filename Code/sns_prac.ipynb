```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values1': [4, 7, 8, 5, 7],
    'Values2': [2, 6, 7, 4, 10],
    'Values3': [3, 8, 5, 7, 6]
}

df = pd.DataFrame(data)

sns.lineplot(x="Category", y="Values1", data=df, label="Values_1", marker="o", color="maroon")
sns.lineplot(x="Category", y="Values2", data=df, label="Values_2", marker="o")

plt.title("Line Plot")
plt.xlabel("Categories")
plt.ylabel("values")
plt.legend()
plt.show()
```

```python
sns.barplot(x="Category", y="Values1", data=df, label="Values_1", color="Teal")

plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("values")
plt.legend()
plt.show()
```

```python
sns.scatterplot(x="Category", y="Values2", data=df, label="Values_2", color="maroon", marker="o")
sns.scatterplot(x="Category", y="Values1", data=df, label="Values_1", color="blue", marker="x")

plt.title("Scatter Plot")
plt.xlabel("Categories")
plt.ylabel("values")
plt.legend()
plt.show()
```

```python
sns.histplot(df["Values1"], color="pink")

plt.title("Hitogram")
plt.xlabel("values")
plt.ylabel("frequencies")
plt.legend()
plt.show()

sns.histplot(df["Values1"], color="pink", bins=2)

plt.title("Customizd Hitogram")
plt.xlabel("values")
plt.ylabel("frequencies")
plt.legend()
plt.show()
```

```python
sns.kdeplot(df["Values2"],fill=True ,color="teal")

plt.title("kde plot of values_2")
plt.xlabel("values_2")
plt.ylabel("density")
plt.legend()
plt.show()
```

```python
data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C','C'],
    'Values': [5, 7, 8, 6, 7, 5, 8, 9, 6, 10]
}

df = pd.DataFrame(data)

sns.boxplot(x="Category", y="Values", data=df)

plt.title("box_plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
sns.displot(df["Values2"], kde=True, color="maroon")

plt.title("dist_plot with kde=true")
plt.xlabel("values")
plt.ylabel("frequencies")
plt.legend()
plt.show()

```
