```python
# sum of all odd numbers
n=int(input("enter number"))
sum_odd=0
for i in range(1,n):
    if i%2!=0:
        sum_odd+=i
print(sum_odd)




# arrays, list
numbers=input("enter numbers")
new=list(numbers)
print(new)

#part(a)
new.reverse()
print(new)

# part(b)
new.sort(reverse=True)
print(new)

# part(c)
numbers=[]
for i in range(3):
    num=int(input(f"enter number {i+1}"))
    numbers.append(num)
print(numbers)

even_count=0
even_numbers=[]
for i in numbers:
    if i % 2 ==0:
        even_numbers.append(i)
        even_count+=1
print("even numbers are", even_count)
print("even list is: ", even_numbers)



# tuple
fruits=("mango", "cherry", "orange")
print("original tuple:", fruits)

new_fruits=list(fruits)
new_fruits.append("banana")
print(new_fruits)

updated_tuple=tuple(new_fruits)
print(updated_tuple)



# Dictionary
info={
    "name":"sara",
    "age":21,
    "marks":700
    }

print(info["marks"])

new_marks=int(input("enter marks"))
info["makrs"]=new_marks
print(new_marks)
print(info)
```
