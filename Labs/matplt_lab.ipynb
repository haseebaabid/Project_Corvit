```python
from numpy._core.fromnumeric import prod
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/matplotlib_practice_dataset.csv')

product_sales=df.groupby("Product")["Units_Sold"].sum()

plt.bar(product_sales.index, product_sales.values, color="purple")


plt.title("bar graph")
plt.xlabel("products")
plt.ylabel("sale")
plt.show()

```

```python
plt.scatter(df["Advertising_Spend"], df["Revenue"], color="maroon")

plt.title("scatter plot")
plt.xlabel("products")
plt.ylabel("sale")
plt.grid(True)
plt.show()
```

```python
plt.hist(df["Units_Sold"], color="pink", bins=7)

plt.title("histogram")
plt.xlabel("units_sold")
plt.ylabel("frequency")
plt.legend("legend")
plt.show()
```

```python
sales_record=df.groupby("Region")["Units_Sold"].sum()

plt.bar(sales_record.index, sales_record.values, color="black")

plt.title("bar chart")
plt.xlabel("region")
plt.ylabel("sale")

plt.show()
```

```python
plt.plot(df["Revenue"].head(10), color="teal")

plt.title("line plot")
plt.xlabel("revenue")
plt.ylabel("frequency")
plt.grid(True)

plt.show()
```
