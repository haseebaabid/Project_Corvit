```python
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd


df = pd.read_csv('/content/drive/MyDrive/diabetes.csv')
df["Diabetic"]=df["Diabetic"].map({0 : "no", 1:"yes"})

average_glucose=df.groupby("Diabetic")["PlasmaGlucose"].mean()
plt.bar(average_glucose.index, average_glucose.values, color="purple")

plt.xlabel("Diabetic")
plt.ylabel("Plasma Glucose")
plt.title("bar graph")
plt.show()
```

This is a pie chart that is showing the percentages of blood pressure in the diabetes dataset.

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd


df = pd.read_csv('/content/drive/MyDrive/diabetes.csv')

import matplotlib.pyplot as plt
bp_bins = [0, 60, 80, 90, 200]
bp_labels = ["Low (<60)", "Normal (60-80)", "Elevated (80-90)", "Hypertensive (90+)"]

df["BP_Group"] = pd.cut(
    df["DiastolicBloodPressure"],
    bins=bp_bins,
    labels=bp_labels
)


bp_counts = df["BP_Group"].value_counts().sort_index()

plt.figure(figsize=(7,7))
plt.pie(
    bp_counts.values,
    labels=bp_counts.index,
    autopct='%1.2f%%',
    startangle=0,
    colors=['skyblue', 'lightgreen', 'orange', 'salmon']
)
plt.title("Diastolic Blood Pressure Distribution")
plt.show()
```
